## Using Decision tree with gridsearch to predict price of diamonds based on their attributes:

### Descriptif of the dataset's features:

* Carat: weight (measured in carat, which is 0.2g)
* Cut: cut quality (in increasing order Fair, Good, Very Good, Premium, Ideal)
* Color: color of the diamond (in increasing quality J, I, H, G, F, E, D), with D being the best and J the worst
* Clarity: a measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))
* Depth: calculated by the height of a diamond, measured from the culet to the table, divided by its average girdle diameter (the image below gives interpretations to all the "diamond-wise" vocabulary).
* Table: The diamond's table size, expressed as a percentage of its diameter.
* Price: the diamond's price. This is the value we wish to predict using a model later on in the notebook
* X: the length of the diamond in mm.
* Y: the width of the diamond in mm.
* Z: the depth of the diamond in mm. 

### Load dataset and create a volume feature: length * width * depth

```python
df = pd.read_csv("diamonds.csv",index_col=0).rename(columns={'x':'length','y':'width','z':'depth2'})

df['volume'] = df['length']*df['width']*df['depth2']
df.head(2)
```

```python
# df.dtypes
```

### Check categorical columns:

```python
for i in df.columns:
    if df[i].dtypes == object:
        print('{:10}: {}'.format(i,list(sorted(df[i].unique()))))
```

### Change categorical value to numbers for plotting below:
We will get ride of them after to use one-hot encoding.

```python
### 1 better 5 worst
cuty = {'Fair':5, 'Good':4, 'Very Good':3, 'Premium':2, 'Ideal':1}
colory =  {'D':1, 'E':2, 'F':3, 'G':4, 'H':5, 'I':6, 'J':7}
clarityy =  {'I1':8,'SI2':7, 'SI1':6, 'VS2':5, 'VS1':4, 'VVS2':3, 'VVS1':2,'IF':1}
  
df['cut-n'] = [float(cuty[i]) for i in df['cut']]
df['color-n'] = [float(colory[i]) for i in df['color']]
df['clarity-n'] = [float(clarityy[i]) for i in df['clarity']]

df.head(2)
```

## Check the price distribution: 
The target we want to predict.

```python
df['price'].hist();
```

Notice, the distribution of the 'price' feature is right-skewed, let's see if log transformation can approximate a normal distribution for the algorithm to  assume normality.

```python
# create a new feature Log_price
df['Log_price'] = np.log(df['price'])
df['Log_price'].hist();
```

Looks like it confirms our hypothesis. Based on this, we could transform it before training our model, however, decision tree does not make any assumption about the distribution so no need.

```python
df.drop('Log_price', axis=1, inplace=True)
```

```python
df1 = df.describe().T
df1
```

### No missing values in features but there are 0s in min, it can't be possible, we are going to keep them and put median of their sub category:

```python
listy = ['length','width','depth2','volume']

for i in listy:
    df[i] = df[i].replace(0,None)
    df[i] = df.groupby(['carat','cut','color','clarity'])[i].transform(lambda x: x.fillna(x.median())) #'depth','table'
df1 = df.describe().T
df1
```

### Visualise all distributions:

```python
df.hist(bins=50, figsize=(17, 12))
plt.show();
```

```python
feat = 'price'
```

## Identify and get ride of outliers:

```python
### Detect and exclude outliers
def detect_outlier(x):
    outliers=[]
    
    threshold_, mean_, std_= 3, np.mean(x), np.std(x)
    
    for y in x:
        z_score= (y - mean_)/std_ 
        if np.abs(z_score) > threshold_:
            outliers.append(y)
    return outliers

print(len(df))

for i in df.columns:
    if df[i].dtypes != object:
        outliers_doy = detect_outlier(df[i])
        df = df[~df[i].isin(outliers_doy)]

print(len(df))
```

```python
cut_cats = ['Fair','Good','Very Good','Premium','Ideal']
col_cats = ['D','E','F','G','H','I','J']
clarity_cats = ['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF']

fig,axes = plt.subplots(1,3,sharey=False) # Prepare the grids
fig.set_size_inches(18,6)                 # Set size of these grids
ax = axes.ravel()

listy = [cut_cats, col_cats, clarity_cats]
listo = ['cut','color','clarity']

# Create plots with looping
for i in range(len(listy)):
    fig = sns.boxplot(y=feat,x=listo[i],hue=listo[i],data=df,palette='husl',order=listy[i],width=0.5,dodge=False, ax=ax[i])
    fig.get_legend().remove()
    ax[i].set_xlabel(listo[i],fontsize=12)
    ax[i].set_ylabel('Diamond Price',fontsize=12);
    ax[i].set_title('Diamond price per {}'.format(listo[i]))

plt.tight_layout();
plt.show();
```

```python
fig,axes = plt.subplots(2,3,sharey=False) # Prepare the grids
fig.set_size_inches(17,7)                 # Set size of these grids
ax = axes.ravel()

listy = ['carat', 'depth', 'table', 'length','width', 'depth2'] #df.columns # ,'volume'

# Create plots with looping
for i in range(len(listy)):
    sns.scatterplot(x=listy[i],y=feat,data=df,ax=ax[i])
    ax[i].set_xlabel(listy[i],fontsize=12)
    ax[i].set_ylabel('Diamond Price',fontsize=12)

plt.tight_layout()
plt.show();
```

```python
# Create a correlation matrix between every pair of attributes
corr = df.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap='Blues', vmax=.3, center=0,annot = True,
            square=True, linewidths=.5, cbar_kws={"shrink": .5});
```

### Drop columns that have no correlation with price:

```python
df.drop(['depth','table','cut-n','color-n','clarity-n'],axis=1,inplace=True)
df.head()
```

## Onehot encoding categorical columns:

```python
# Create dummies
print(len(df))

df_onehot = pd.DataFrame()
for i in ['cut','color','clarity']:
    df_onehot = pd.concat([df_onehot,pd.get_dummies(df[i],prefix=i)],axis = 1)
    
df = pd.concat([df,df_onehot],axis=1)

df.drop(['cut','color','clarity'],axis=1,inplace=True) 
print(len(df))
df.head()
```

```python
print(df.isnull().any(),np.all(np.isfinite(df)))
```

## Build decision tree model:

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier

# Evaluation metrics
from sklearn.metrics import mean_absolute_error,mean_squared_error,mean_squared_error
from sklearn.model_selection import KFold

# Grid-search
from sklearn.model_selection import GridSearchCV

from sklearn.decomposition import PCA
```

### Split into train test sets:

```python
X_train, X_test, y_train, y_test = train_test_split(df[[i for i in df.columns if i != feat]],df[[feat]],test_size=0.2)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
for i in [X_train, X_test, y_train, y_test]:
    i = i.reset_index(drop=True, inplace=True)
```

### Set parameters:

```python
# Hyperparameters:
max_depth = [int(x) for x in np.linspace(10,30,num=5)]
max_depth.append(None)
min_samples_split = [30,40,50]
min_samples_leaf = [6,7]

# parameters:
param_grid = {'max_depth':max_depth,
              'min_samples_split':min_samples_split,
              'min_samples_leaf':min_samples_leaf}
```

### Model of Decision Tree Regression and GridSearch

```python
# Model of Decision Tree Regression
tree= DecisionTreeRegressor()
# Conduct 10-fold CV
kfold = KFold(n_splits=10,random_state=42,shuffle=True)
# Grid-search
grid_search_tree = GridSearchCV(tree, # Model
                                param_grid = param_grid, # Hyperparameters
                                scoring = ['neg_mean_absolute_error','neg_root_mean_squared_error'], # Scores
                                refit = 'neg_root_mean_squared_error', # Specific score for tunning
                                cv = kfold, # Cross validation method
                                return_train_score = True) # Return any scores in training data CV
```

```python
# Hyperparameters tunning with training data
grid_search_tree.fit(X_train,y_train)
```

```python
# Show scores in fitting models
pd.DataFrame(grid_search_tree.cv_results_).head(2)
```

```python
# Best model
print('Best hyperparameters : ',grid_search_tree.best_params_,'\n')
print('Best evaluation : ',grid_search_tree.best_score_,'\n')
print('Best model of Decision Tree: ',grid_search_tree.best_estimator_,'\n')
```

```python
tree_model = grid_search_tree.best_estimator_
tree_model.fit(X_train,y_train)
```

### Error Metrics on training and testing sets:

```python
print('Training set:  MSE: {:20}, RMSE: {:20}, MAE: {:20}, Corr: {:20}'.format(mean_squared_error(y_true=y_train,y_pred=tree_model.predict(X_train)),
                                                                               np.sqrt(mean_squared_error(y_true=y_train,y_pred=tree_model.predict(X_train))),
                                                                               mean_absolute_error(y_true=y_train,y_pred=tree_model.predict(X_train)),
                                                                               np.corrcoef(tree_model.predict(X_train),y_train[feat].values)[0,1]))

print('Test set:      MSE: {:20}, RMSE: {:20}, MAE: {:20}, Corr: {:20}'.format(mean_squared_error(y_true=y_test,y_pred=tree_model.predict(X_test)),
                                                                               np.sqrt(mean_squared_error(y_true=y_test,y_pred=tree_model.predict(X_test))),
                                                                               mean_absolute_error(y_true=y_test,y_pred=tree_model.predict(X_test)),
                                                                               np.corrcoef(tree_model.predict(X_test),y_test[feat].values)[0,1]))
```

- RMSE is the standard deviation of the residuals (prediction errors). Residuals are a measure of how far from the regression line data points are; RMSE is a measure of how spread out these residuals are. In other words, it tells you how concentrated the data is around the line of best fit.

- The MAE (mean absolute error) is easier to interpret: it's literally the mean of the absolute value of the error.

### Mean absolute percentage error gives an easy to interpret percentage showing how off the predictions are:

```python
def mape(targets, predictions): 
    targets, predictions = np.array(targets), np.array(predictions)
    return str(np.round(np.mean(np.abs((targets - predictions) / targets)) * 100,2))+'%'
```

```python
print('Mean absolute percentage error:  Training: {:4}, Testing: {:4}'.format(mape(y_train[feat],tree_model.predict(X_train)),
                                                                              mape(y_test[feat],tree_model.predict(X_test))))
```
